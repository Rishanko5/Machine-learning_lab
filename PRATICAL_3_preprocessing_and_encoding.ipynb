```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
#load dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
print(df.isnull().sum())
```

```python
df.loc[5:12, 'sepal width (cm)'] = np.nan
```

```python
print(df['sepal width (cm)'].head(20))
```

```python
mean_sepal_width = df['sepal width (cm)'].mean() #mean
mean_sepal_width
```

```python
print(df.isnull().sum())
```

```python
df['sepal width (cm)'] = df['sepal width (cm)'].fillna(mean_sepal_width)
```

```python
print(df['sepal width (cm)'].isnull().sum())
```

```python
#box plot
sns.boxplot(x=df['sepal width (cm)'])
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, orient='h')
plt.title('Box Plot of All Features')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

df['species_encoded'] = label_encoder.fit_transform(df['species'])

display(df.head())
```
